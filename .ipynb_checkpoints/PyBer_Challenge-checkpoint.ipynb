# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"
```

```python
# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head(10)
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
total_rides_by_citytype = pyber_data_df.groupby(["type"]).count()["ride_id"]
#total_rides_urban = pyber_data_df[pyber_data_df["type"]=="Urban"].sum()
#total_rides_urban
total_rides_by_citytype
```

```python
# 2. Get the total drivers for each city type
total_drivers_by_citytype = pyber_data_df.groupby(["type"]).sum()["driver_count"]
total_drivers_by_citytype
```

```python
#  3. Get the total amount of fares for each city type
total_fares_by_citytype = pyber_data_df.groupby(["type"]).sum()["fare"]
total_fares_by_citytype
```

```python
#  4. Get the average fare per ride for each city type. 
average_fares_by_citytype = pyber_data_df.groupby(["type"]).mean()["fare"]
average_fares_by_citytype
```

```python
#  4. Alternative method get the average fare per ride for each city type. 
average_fares_per_ride= total_fares_by_citytype/total_rides_by_citytype
average_fares_per_ride
```

```python
# 5. Get the average fare per driver for each city type. 
average_fares_perDriver= total_fares_by_citytype/total_drivers_by_citytype
average_fares_perDriver
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df=pd.DataFrame(
    {"Total Rides": total_rides_by_citytype,
     "Total Drivers": total_drivers_by_citytype,
     "Total Fares": total_fares_by_citytype,
     "Average Fare per Ride": average_fares_per_ride,
     "Average Fare per Driver": average_fares_perDriver   
})
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
#  8. Format the columns.
pyber_summary_df["Total Fares"]=pyber_summary_df["Total Fares"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Driver"]= pyber_summary_df["Average Fare per Driver"].map("${:,.2f}".format)

pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 0. not 1. Read the merged DataFrame
pyber_data_df
```

```python
# 1 not 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
new_pyber_data_df = pd.DataFrame(pyber_data_df.groupby(["type", "date"]).sum()["fare"])
new_pyber_data_df
```

```python
# 2 not 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
new_pyber_data_df = new_pyber_data_df.reset_index()
new_pyber_data_df.head()
```

```python
# 3 not 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
new_pyber_pivot_df=new_pyber_data_df.pivot(index='date', columns='type', values='fare')
new_pyber_pivot_df
```

```python
# 4 not 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
new_pyber_pivot_locdates_df = new_pyber_pivot_df.loc[(new_pyber_pivot_df.index > "2019-01-01") & (new_pyber_pivot_df.index <= "2019-04-29")]
new_pyber_pivot_locdates_df
```

```python
# 5 not 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
new_pyber_pivot_locdates_df.index = pd.to_datetime(new_pyber_pivot_locdates_df.index)
```

```python
# 6 not 7. Check that the datatype for the index is datetime using df.info()
new_pyber_pivot_locdates_df.info()
```

```python
# 7 not 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
Weekly_fares_df = new_pyber_pivot_locdates_df.resample("W").sum()
Weekly_fares_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 
ax = Weekly_fares_df.plot(figsize=(16, 6))
lgnd = plt.legend(loc="center", title="type")
#remove index name from the dataframe
Weekly_fares_df.index.name = None
# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
ax.set_ylabel("Fare($USD)")
ax.set_title("Total Fare by City Type")
#ax.set_xticks(x_label)
plt.savefig("Resources/PyBer_fare_summary.png")
plt.show()

```
